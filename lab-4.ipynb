**1) to add,subtract,multiply,divide arguments element-wise**

```python
array1 =np.arange(5)
print("array1:",array1)

sum = array1+4
print("result array with elements plus by 4",sum)

sub = sum-1
print("result array with elements minus by 1",sub)

mul = sub*5
print("result array with elements multipy by 4",mul)

div = mul/2
print("result array with elements divide by 4",div)
```
